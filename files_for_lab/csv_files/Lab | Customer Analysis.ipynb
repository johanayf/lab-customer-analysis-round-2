```python
#Show a plot of the response rate by the sales channel.
def response_rate(data):
    rr=round(data['response'].value_counts(normalize=True).Yes,3)
    return rr
data.groupby(['sales_channel']).apply(response_rate).plot(kind='bar')
```

```python
data['total_claim_amount'].describe()
```

```python
#Show a plot of the response rate by the total claim amount.
#I separated the total claim amount into 4 categories by using percentile 0-25, 25-50, 50-75, 75-100. 
bins = [0, 25, 50, 75, 100]
for i in range(len(bins)):
    if i != len(bins)-1:
        lower_bound = np.percentile(data['total_claim_amount'], bins[i])
        upper_bound = np.percentile(data['total_claim_amount'], bins[i+1])
        data.loc[((data['total_claim_amount'] >= lower_bound) & (data['total_claim_amount'] <= upper_bound)), 'total_claim_amount_binned'] = "{} to {}".format(int(lower_bound), int(upper_bound))
```

```python
data['total_claim_amount_binned'].value_counts()
```

```python
data.groupby(['total_claim_amount_binned']).apply(response_rate).plot(kind='bar')
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_num.isnull().sum()
```

```python
#check if there is infinite value
def null_clean(x):
    if np.isfinite(x):
        return x
    else:
        return np.NAN
```

```python
#Run the function with the columns with null value
df_num['months_since_last_claim'] = list(map(null_clean, df_num['months_since_last_claim']))
df_num['number_of_open_complaints'] = list(map(null_clean, df_num['number_of_open_complaints']))
```

```python
#Replacing the null value with mean
#Question: why are there warming? 
df_num['months_since_last_claim'] =df_num['months_since_last_claim'].fillna(np.mean(df_num['months_since_last_claim']))
df_num['number_of_open_complaints'] =df_num['number_of_open_complaints'].fillna(np.mean(df_num['number_of_open_complaints']))
```

```python
#X,y Split
Y = df_num['total_claim_amount']
X = df_num.drop('total_claim_amount', axis= 1)
lm = linear_model.LinearRegression()
lm.fit(X,Y)
```
